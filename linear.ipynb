```python
import numpy as np  # You need to import the numpy library to use it

class Linear_Regression:
    def __init__(self):
        self.alpha = 0.01
        self.history = []
        self.epochs = 100
        self.n_features = 0
        self.n_examples = 0
        self.W = None
        
    def fit(self, x, y):
        self.X = x
        self.Y = y
        self.n_features = x.shape[1] 
        self.n_examples = x.shape[0]
        self.col = np.ones((self.n_examples, 1))  
        self.X = np.append(self.X, self.col, axis=1)  
        self.W = np.random.rand(self.n_features+1, 1) 

    def set_learning_rate(self, alpha):
        self.alpha = alpha
    def set_epochs(self, epochs):
        self.epochs = epochs
    def train(self):
        for e in range(self.epochs):
            Yh = np.dot(self.X,self.W)
            cost = np.sum((Yh-self.Y)**2)/(self.n_examples*2)
            self.history.append(cost)
            gradient =  np.dot((Yh-self.Y).T,self.X).T /(self.n_examples)
            self.W -= self.alpha*gradient
    def predict(self,x):
        return self.W[0,0]*x + self.W[1,0]
    def plot_history(self):
        import matplotlib.pyplot as plt
        plt.plot(np.arange(0,self.epochs,1),self.history)


```

```python
noise =np.random.rand(100,1)
noise/=np.max(noise)
x = np.expand_dims(np.arange(0,1,0.01),1) - noise/10
y = np.expand_dims(np.arange(0,1,0.01),1)
model = Linear_Regression()
model.fit(x, y)  
```

```python
model.set_epochs(1000)
model.set_learning_rate(0.01)
model.train()
```

```python
model.predict(2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,model.W[0,0]*x+model.W[1,0],c='r')
```

```python
model.plot_history()
```
